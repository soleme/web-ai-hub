## '함수형'의 의미
- 절차형 프로그래밍
- 객체지향형 프로그래밍
- 함수형 프로그래밍

```python
def add(a, b):
    return a + b
print(add(1, 3))
print(type(add))
```

```python
number = 1
print(number)
print(type(number))
```

```python
def add(a, b):
    return a + b

def minus(a, b):
    return a - b

def func(f, a, b):
    return f(a, b)

print(func(add, 3, 2))
print(func(minus, 3, 2))
```

```python
# lambda
add = lambda a, b: a + b
minus = lambda a, b: a - b
def func(f, a, b):
    return f(a, b)

print(func(add, 3, 2))
print(func(minus, 3, 2))
print(func(lambda a, b: a * b, 3, 2))
print(func(lambda a, b: a / b, 3, 2))
```

```python
def get_func(what='add'):
    if what == 'minus':
        return lambda a, b: a - b
    return lambda a, b: a + b

print(get_func()(1, 3))
print(get_func('add')(1, 3))
print(get_func('minus')(1, 3))
```

## 보스턴 집값 예측

```python
import tensorflow as tf
import pandas as pd

boston = pd.read_csv("https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/boston.csv")
print(boston.columns)
x_boston = boston.drop(['medv'], axis=1)
y_boston = boston[['medv']]
print(x_boston.shape, y_boston.shape)
```

```python
X = tf.keras.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.Model(X, Y)
model.compile(optimizer='adam', loss=tf.keras.losses.mse)
```

```python
x_train, x_test = x_boston[:400], x_boston[400:]
y_train, y_test = y_boston[:400], y_boston[400:]
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
result = model.fit(x_train, y_train, batch_size=128, epochs=500, validation_split=0.2)
```

### Train / Validation / Test 
- train: 학습용
- validation: 모의고사용
- test: 최종면접

```python
x_train, x_test = x_boston[:400], x_boston[400:]
y_train, y_test = y_boston[:400], y_boston[400:]
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
result = model.fit(x_train, y_train, batch_size=128, epochs=500, validation_split=0.2, verbose=True)
```

```python
model.evaluate(x_test, y_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(result.history['loss'][10:])
plt.plot(result.history['val_loss'][10:])
plt.legend(['loss', 'val_loss'])
plt.show()
```
